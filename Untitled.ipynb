```python
import pandas as pd

file_load = "Temp_avg_Year.csv"

temp_avg_year = pd.read_csv(file_load)

temp_avg_year.head()

temp_avg_year.info()
```

```python
#convert celsius to farneheight 
#formula=(0°C × 9/5) + 32

to_faren = (temp_avg_year["Temperature - (Celsius)"] * (9/5)) + 32

to_faren
```

```python
#append temps in deg farenheight to df 

temp_avg_year = temp_avg_year.append(to_faren)

temp_avg_year
```
